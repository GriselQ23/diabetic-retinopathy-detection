```python
import os
import cv2
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import load_model

# Cargar el modelo entrenado
modelo_entrenado = load_model('/content/drive/Shareddrives/Darwin Project Datasets II/Modelos/modelo.h5')

# Directorio de imágenes
directorio_imagenes = '/tmp/test'
clases = os.listdir(directorio_imagenes)

# Almacenar las imágenes y las etiquetas
imagenes = []
etiquetas = []

# Cargar las imágenes y las etiquetas
for clase in clases:
    ruta_clase = os.path.join(directorio_imagenes, clase)
    for imagen_nombre in os.listdir(ruta_clase):
        ruta_imagen = os.path.join(ruta_clase, imagen_nombre)
        imagen = cv2.imread(ruta_imagen)
        #imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)  # Convertir a RGB si es necesario
        imagen = cv2.resize(imagen, (224, 224))  # Redimensionar la imagen si es necesario
        imagenes.append(imagen)
        etiquetas.append(clase)

# Convertir las listas a matrices numpy
imagenes = np.array(imagenes)
etiquetas = np.array(etiquetas)

# Encode the labels as numeric values
label_encoder = LabelEncoder()
etiquetas = label_encoder.fit_transform(etiquetas)

# Realizar las predicciones
predicciones = modelo_entrenado.predict(imagenes)

# Obtener las etiquetas predichas
etiquetas_predichas = np.argmax(predicciones, axis=1)

# Create confusion matrix
matriz_confusion = confusion_matrix(etiquetas, etiquetas_predichas)

# Get class labels from label encoder
class_labels = label_encoder.classes_

# Plot confusion matrix with colors
plt.imshow(matriz_confusion, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Matriz de Confusión')
plt.colorbar()

tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels, rotation=45)
plt.yticks(tick_marks, class_labels)

thresh = matriz_confusion.max() / 2.
for i in range(matriz_confusion.shape[0]):
    for j in range(matriz_confusion.shape[1]):
        plt.text(j, i, format(matriz_confusion[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if matriz_confusion[i, j] > thresh else "black")

plt.ylabel('Etiqueta Verdadera')
plt.xlabel('Etiqueta Predicha')
plt.tight_layout()
plt.show()
```
